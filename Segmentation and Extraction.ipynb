```python
def get_paragraphs(lyrics):
    return lyrics.split(splitter)    

# Global Variables
splitter = "\n  \n"

from rake_nltk import Rake
import pandas as pd
```

```python
# Pre-processing
Songs = pd.read_csv("songdata.csv")
Songs.set_index("song",inplace=True)
import time
start_time = time.time()
Songs = Songs["text"].apply(get_paragraphs)
print("My program took", time.time() - start_time, "seconds to run")
Songs.head()
```

```python
for j in range(len(Songs.head(1))):
    for i in range(len(Songs.head(1)[j])):
        print(str(Songs.head(1)[j][i]) + "\n")
```

```python
for list in Songs.iloc[22]:
    text = list
    print("Sentence: " +str(text.split("\n")[0]) )

    rake = Rake()
    rake.extract_keywords_from_text(text.split("\n")[0])
    print(rake.get_ranked_phrases_with_scores())
    print("\n")
```
