```python
import requests
from IPython.display import Image, display

def getImgUrls(keywords):
    urls=[]
    r = requests.get('https://pixabay.com/api/?key=6713313-c85a6d8e3f6fe1a13b85f7d78&q='+"+".join(keywords)+'&image_type=photo&page=1&per_page=3')
    for img in r.json()['hits']:
        urls.append(img['webformatURL'])
    return urls

keywords=['tears']
urls=getImgUrls(keywords)
for url in urls:
    print(url)
    display(Image(url,height=400,width=1000,embed=True))
```
